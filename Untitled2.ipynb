**Parse Transcripts & Build a Metadata Table**

```python
import glob

# Path to dev-clean audio folder
data_path = '/content/drive/MyDrive/speech_to_text_project/data/audio/dev-clean'

# Store all transcript entries
metadata = []

for txt_file in glob.glob(f'{data_path}/**/*.trans.txt', recursive=True):
    with open(txt_file, 'r') as f:
        for line in f:
            parts = line.strip().split(' ', 1)
            file_id, transcript = parts[0], parts[1]
            speaker_id = file_id.split('-')[0]
            chapter_id = file_id.split('-')[1]
            audio_path = os.path.join(data_path, speaker_id, chapter_id, f'{file_id}.flac')
            metadata.append((audio_path, transcript))

# Convert to DataFrame
import pandas as pd

df = pd.DataFrame(metadata, columns=['audio_path', 'transcript'])
df.head()
```

**Extract MFCCs in Batches**

```python
import numpy as np
from tqdm import tqdm

# Function to extract and flatten MFCCs for simple use (e.g., for classical ML models)
def extract_mfcc_features(file_list, n_mfcc=13):
    features = []
    errors = []

    for path in tqdm(file_list):
        try:
            y, sr = librosa.load(path, sr=16000)
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
            mfcc_mean = np.mean(mfcc.T, axis=0)  # Mean over time axis
            features.append(mfcc_mean)
        except Exception as e:
            errors.append((path, str(e)))
            features.append(np.zeros(n_mfcc))  # Placeholder for failed files

    return np.array(features), errors

# Run it
mfcc_features, errors = extract_mfcc_features(df['audio_path'])

# Save results into DataFrame
df_mfcc = df.copy()
df_mfcc['mfcc'] = list(mfcc_features)
df_mfcc.head()
```

**Define Beam Search Decoder**

```python
import numpy as np

def beam_search(hmm_model, mfcc, char_encoder, beam_width=5):
    n_samples = mfcc.shape[0]
    n_states = hmm_model.n_components
    log_probs = hmm_model._compute_log_likelihood(mfcc)  # (n_samples, n_states)

    beams = [([], 0.0)]  # (sequence, log_prob)

    for t in range(n_samples):
        new_beams = []
        for seq, score in beams:
            for s in range(n_states):
                new_seq = seq + [s]
                new_score = score + log_probs[t, s]
                new_beams.append((new_seq, new_score))

        # Keep top N beams
        new_beams = sorted(new_beams, key=lambda x: x[1], reverse=True)[:beam_width]
        beams = new_beams

    # Choose best beam
    best_seq = beams[0][0]
    decoded_chars = char_encoder.inverse_transform(best_seq)
    decoded_text = ''.join(decoded_chars)

    return decoded_text
```
